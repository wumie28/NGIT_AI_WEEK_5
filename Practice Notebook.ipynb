## 2. Reading and Writing Files (File I/O)

File handling is a crucial part of any programming language. Python provides built-in functions to create, read, update, and delete files.

### The `open()` Function
The key function for working with files in Python is `open()`. It takes two main parameters: `filename` and `mode`.

### File Modes

| Mode | Description |
| :--- | :--- |
| `'r'` | **Read** - Default value. Opens a file for reading, error if the file does not exist. |
| `'a'` | **Append** - Opens a file for appending, creates the file if it does not exist. |
| `'w'` | **Write** - Opens a file for writing, creates the file if it does not exist. Overwrites existing content! |
| `'x'` | **Create** - Creates the specified file, returns an error if the file exists. |

### The `with` Statement
It is best practice to use the `with` statement when working with files. It automatically closes the file for you when you are done, even if an error occurs. This prevents memory leaks and file corruption.

````python
# --- Writing to a file ---
filename = "sample_data.py"

# 'w' mode overwrites any existing content
with open(filename, "w") as file:
    file.write("'My name is Python.\n'")
    file.write("'Python makes file handling very simple.\n'")
    file.write("'We are learning about the 'w' mode.\n'")

print(f"Successfully wrote to '{filename}'.")
````

````python
# --- Appending to a file ---
# 'a' mode adds content to the end of the file without deleting existing content
with open(filename, "a") as file:
    file.write("This line is appended using 'a' mode.\n")
    file.write("Notice how the previous content is still there!\n")

print(f"Successfully appended to '{filename}'.")
````

## 3. Modules and Packages

A **module** is simply a file containing Python code. A **package** is a directory of modules containing a special `__init__.py` file.

### Why use modules?
- **Reusability**: Write code once, use it everywhere.
- **Organization**: Keep your code clean and manageable.
- **Namespace**: Avoid variable name conflicts.

### Importing Modules
You can import modules in several ways:
1. `import module_name` (Access via `module_name.function()`)
2. `import module_name as alias` (Access via `alias.function()`)
3. `from module_name import function_name` (Access directly via `function_name()`)
4. `from module_name import *` (Imports everything - generally discouraged)

````python
import math
import random

my_list = [1,2,3,4,5,6,7,8]

random.choice(my_list)
math.sqrt(25)
````

````python
# --- Using Built-in Modules ---

import math
import random
import datetime as dt

# Math module
print(f"Square root of 25: {math.sqrt(25)}")
print(f"Factorial of 5: {math.factorial(5)}")

# Random module
print(f"Random number (1-100): {random.randint(1, 100)}")
choices = ['Apple', 'Banana', 'Cherry']
print(f"Random choice: {random.choice(choices)}")

# Datetime module
now = dt.datetime.now()
print(f"Current date and time: {now}")
print(f"Current year: {now.month}")
````

## 4. NumPy (Numerical Python)

NumPy is the fundamental package for scientific computing in Python. It provides support for arrays (matrices), which are much faster and more efficient than Python lists.

### Why NumPy?
- **Speed**: NumPy arrays are stored in contiguous memory locations.
- **Functionality**: Optimized built-in mathematical functions.
- **Convenience**: Easy to perform operations on entire arrays at once (broadcasting).

### Most Used NumPy Methods

| Method | Description | Example |
| :--- | :--- | :--- |
| `np.array(list)` | Creates a NumPy array from a python list | `arr = np.array([1, 2, 3])` |
| `np.zeros(shape)` | Creates an array filled with zeros | `np.zeros((2, 3))` |
| `np.ones(shape)` | Creates an array filled with ones | `np.ones((2, 3))` |
| `np.arange(start, stop, step)` | Creates an array with a range of values | `np.arange(0, 10, 2)` |
| `np.linspace(start, stop, num)` | Creates an array with `num` evenly spaced values | `np.linspace(0, 1, 5)` |
| `arr.reshape(new_shape)` | Changes the shape of an array | `arr.reshape(2, 3)` |
| `arr.shape` | Attribute returning the dimensions of the array | `(2, 3)` |
| `np.sum(arr)` | Calculates sum of all elements | `np.sum(arr)` |
| `np.mean(arr)` | Calculates average of all elements | `np.mean(arr)` |
| `np.max(arr)` / `np.min(arr)` | Finds maximum / minimum value | `np.max(arr)` |
| `np.random.rand(d0, d1)` | Generates random values between 0 and 1 | `np.random.rand(3, 2)` |

````python
import numpy as np

# --- 1. Array Creation ---
arr_1d = np.array([1, 2, 3, 4, 5])
print(f"1D Array: {arr_1d}")

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(f"\n2D Array:\n{arr_2d}")
print(f"Shape of 2D Array: {arr_2d.shape}")

# --- 2. Built-in Creation Functions ---
zeros = np.zeros((3, 3))
print(f"\nZeros Array:\n{zeros}")

range_arr = np.arange(0, 20, 2) # Start at 0, stop before 20, step 2
print(f"\nRange Array: {range_arr}")

# --- 3. Array Operations ---
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Element-wise addition (Broadcasting)
print(f"\na + b: {a + b}")
print(f"a * 2: {a * 2}")

# Statistics
print(f"\nSum: {np.sum(a)}")
print(f"Mean: {np.mean(a)}")
print(f"Max: {np.max(a)}")
````

### NumPy Data Cleaning

Data cleaning is essential. In NumPy, we often deal with missing values represented as `NaN` (Not a Number).

Common tasks:
- Identifying `NaN` values: `np.isnan(arr)`
- Removing `NaN` values: Using boolean indexing
- Replacing `NaN` values: `np.nan_to_num(arr)` or custom logic

````python
# --- Handling Missing Values in NumPy ---

# Create an array with NaN (Not a Number)
data_arr = np.array([1.0, 2.0, np.nan, 4.0, 5.0, np.nan])
print(f"Original Array with NaNs: {data_arr}")

# Check for NaNs
has_nans = np.isnan(data_arr)
print(f"Is NaN mask: {has_nans}")

# Remove NaNs
clean_arr = data_arr[~has_nans] # ~ means 'not'
print(f"Array without NaNs: {clean_arr}")

# Replace NaNs with a value (e.g., 0)
replaced_arr = np.nan_to_num(data_arr, nan=0.0)
print(f"Array with NaNs replaced by 0: {replaced_arr}")

# Replace NaNs with the mean of valid elements
mean_val = np.nanmean(data_arr) # nanmean calculates mean ignoring NaNs
data_arr[np.isnan(data_arr)] = mean_val
print(f"Array with NaNs replaced by Mean ({mean_val}): {data_arr}")
````

### Pandas Data Cleaning

Data cleaning is a critical step in data analysis. Real-world data is often messy, containing missing values, duplicates, or incorrect data types.

Common tasks:
- **Handling Missing Data**: `dropna()`, `fillna()`
- **Removing Duplicates**: `drop_duplicates()`
- **Data Type Conversion**: `astype()`
- **Renaming Columns**: `rename()`

````python
# --- Pandas Data Cleaning Examples ---

# Create a messy DataFrame
messy_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', None],
    'Age': [25, np.nan, 30, 25, 22],
    'Score': ['85', '90', 'NaN', '85', '70'] # Score is string type!
}
df_messy = pd.DataFrame(messy_data)

print("--- Messy DataFrame ---")
print(df_messy)

# 1. Check for missing values
print("\n--- Missing Values Count ---")
print(df_messy.isnull().sum())

# 2. Remove duplicates
df_clean = df_messy.drop_duplicates()
print("\n--- After Removing Duplicates ---")
print(df_clean)

# 3. Handling Missing Values
# Fill Name with 'Unknown'
df_clean['Name'] = df_clean['Name'].fillna('Unknown')
# Fill Age with mean age
mean_age = df_clean['Age'].mean()
df_clean['Age'] = df_clean['Age'].fillna(mean_age)

print("\n--- After Filling Missing Values ---")
print(df_clean)

# 4. Data Type Conversion
# 'Score' is object (string) because of 'NaN' string. Let's fix it.
# First, coerce errors to NaN (turn string 'NaN' to actual NaN)
df_clean['Score'] = pd.to_numeric(df_clean['Score'], errors='coerce')

# Now fill the new NaN in score with 0 or mean
df_clean['Score'] = df_clean['Score'].fillna(0)

print("\n--- Final Cleaned DataFrame ---")
print(df_clean)
print(df_clean.info())
````
